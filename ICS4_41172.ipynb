```python
import numpy as np
import math
```

```python
# function to calculate gcd of two numbers
def gcd(a, b):
  if b == 0:
    return a
  else:
    return gcd(b, a % b)
```

```python
# function to calculate the value of exponent e
def select_e(n, phi):
  e = 2
  while e < phi:
    if gcd(e, phi) == 1:
      break
    else:
      e += 1
  return e
```

```python
# function to encrypt plain data
def encrypt_data(data, public_key):
  e = public_key[1]
  cipher = math.pow(data, e) % n
  return cipher
```

```python
# function to decrypt cipher text
def decrypt_data(cipher, private_key):
  d = private_key[1]
  original = math.pow(cipher, d) % n
  return original
```

```python
p = 3
q = 7
n = p * q
phi = (p - 1) * (q - 1)
e = select_e(n, phi)
```

```python
# generate the public key
public_key = (n, e)
print("Public Key: ", public_key)
```

```python
# generating private key
k = 2
d = ((k * phi) + 1) / e
private_key = (n, d)
print("Private Key: ", private_key)
```

```python
# encrypting - decrypting the data
data = 12
cipher = encrypt_data(data, public_key)
original = decrypt_data(cipher, private_key)

print("Data: ", data)
print("Cipher Text (After encryption): ", cipher)
print("Plain Text (After decryption): ", original)
```

**Implementation of Diffie-Hellman key exchange**

```python
import math
import random
```

```python
def generate(p, g):
  print("Diffie Hellman Key Exchange")

  # generate two random numbers a, b
  a = random.randint(2, 1000)
  b = random.randint(2, 1000)

  print("Modulus chosen: ", p)
  print("Base chosen: ", g)
  print("Number selected by A: ", a)
  print("Number selected by B: ", b)

  print("Calculating Shared Keys for both A and B: ")
  x = pow(g, a) % p
  y = pow(g, b) % p

  print("A's calculated value: ", x)
  print("B's calculated value: ", y)
  print("Exchange of the generated keys takes place")

  ka = pow(y, a) % p
  kb = pow(x, b) % p

  print("A's secret key: ", ka)
  print("B's secret key: ", kb)

```

```python
# public keys available
p = 95
g = 23
generate(p, g)
```

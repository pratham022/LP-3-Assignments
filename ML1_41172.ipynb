The following table shows the results of a recently conducted study on the correlation of the number of hours spent driving with the risk of developing acute backache. 
Find the equation of the best fit line for this data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
X_driving_hours = [10, 9, 2, 15, 10, 16, 11, 16]
y_risk_score = [95, 80, 10, 50, 45, 98, 38, 93]
```

```python
# Plot the graph of driving hrs vs risk score
plt.plot(X_driving_hours, y_risk_score, 'o')
```

Finding coefficients b0, b1 that satisfy the equation y = b1x + b0 for given dataWhatsApp Image 2022-05-11 at 8.29.55 PM.jpeg

```python
# helper functions
def mean(values):
  return sum(values) / float(len(values))
def variance(values, mean):
  return sum([(x - mean)**2 for x in values])
def covariance(x, mean_x, y, mean_y):
  covar = 0.0
  for i in range(len(x)):
    covar += (x[i] - mean_x) * (y[i] - mean_y)
  return covar
```

```python
# find the coefficients
def find_coefficients(x, y):
  mean_x = mean(x)
  mean_y = mean(y)
  b1 = covariance(x, mean_x, y, mean_y) / variance(x, mean_x)
  b0 = mean_y - b1 * mean(x)
  return [b0, b1]
```

```python
# call the above functions from here
b0, b1 = find_coefficients(X_driving_hours, y_risk_score)
print(f'Line Equation: y = {round(b1, 2)} * x + {round(b0, 2)}')
```

```python
# plot the line
def plot_graph(slope, intercept):
  axes = plt.gca()
  x_vals = np.array(axes.get_xlim())
  y_vals = intercept + slope * x_vals
  plt.plot(x_vals, y_vals, '-')
```

```python
plt.plot(X_driving_hours, y_risk_score, 'o')
plot_graph(b1, b0)
```

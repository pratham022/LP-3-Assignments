```python
import math
import random
```

```python
def get_generator_point(a, b):
  if (4*(a**3)) + (27*(b**2)) != 0:
    x = 1
    while True:
      rhs = x**3 + a*x + b
      y = int(math.sqrt(rhs))
      lhs = y**2
      if lhs == rhs:
        return [x, y]
      else:
        x += 1
  else:
    print("Select another values for a and b")
    return [-1, -1]
```

```python
# set values for a and b to be used in the equation
a = 2
b = 3

# set the private keys for A and B
private_a = 13
private_b = 15

# prepare the generator point
generator = get_generator_point(a, b)
print("Generator point: ", generator)

# take the plain text input
m = 12

# generate public key for A
public_a = [private_a*generator[0], private_a*generator[1]]

# generate public key for B
public_b = [private_b*generator[0], private_b*generator[1]]

print("Public Key of A: ", public_a)
print("Public Key of B: ", public_b)

# set k to a random integer between 0 to 10
k = random.randint(0, 10)

# encrypt the plaintext into cipher text
c1 = k * (generator[0] + generator[1])

c2 = m + (k * (public_b[0] + public_b[1]))

cipher = [c1, c2]
print("Cipher Text: ", cipher)

# decrypt the ciphertext into plaintext
# first calculate value for r
r = private_b * c1
original_msg = c2 - r
print("Original Message: ", original_msg)
```
